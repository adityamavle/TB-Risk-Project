```python
anova_cols=['Xpert_mtb_detection_result','Death','age']
```

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols
```

```python
model = ols('age ~ C(Xpert_mtb_detection_result) + C(Death) +\
C(Xpert_mtb_detection_result):C(Death)',
            data=dtset2[anova_cols]).fit()
result = sm.stats.anova_lm(model, type=2)
```
